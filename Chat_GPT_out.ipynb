```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/ChatGpt out/gpt output.csv')
```

```python
df.head()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
gold = df['Label'].to_numpy()
fun = df['ChatGpt is funny'].to_numpy()
pun = df['ChatGpt is pun'].to_numpy()
```

```python
accuracy_score(gold, fun)
```

```python
accuracy_score(gold, pun)
```

```python
conf_fun = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(gold, fun), display_labels=[0,1])
conf_fun.plot()
```

```python
conf_pun = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(gold, pun), display_labels=[0,1])
conf_pun.plot()
```

```python
# missclassified instances
for i, row in df.iterrows():
  if row['Label'] == row['ChatGpt is funny'] == row['ChatGpt is pun']:
    df = df.drop(i)
df.shape
```
